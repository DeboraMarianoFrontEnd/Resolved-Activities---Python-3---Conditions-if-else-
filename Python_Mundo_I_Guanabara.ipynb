```python
#18 - O correto é fazer:

import math

angulo = float(input('Digite o angulo: '))

print(f'O seno do seu angulo é de {math.sin(math.radians(angulo))}, já o cosseno é de {math.cos(math.radians(angulo))} e a tangente é de {math.tan(math.radians(angulo))}')
```

```python
#a)

nome = str(input('Digite o seu nome: '))
print(nome.upper())
```

```python
#b)
nome = str(input('Digite o seu nome: '))
print(nome.lower())
```

```python
#C) .strip = tira os espaços inúteis - Veja que quando coloquei o meu nome, adicionei espaço antes de colocar meu primeiro nome e o strip tirou o espaço e contabilizou
#somente as letras e o espaço entre a debora e o sobrenome.

#tenho que colocar o len para fazer a contagem das letras, mas se colocasse somente ele, ele ia contar todos os espaços.

nome = str(input('Digite o seu nome: ')).strip()

print(len(nome) - nome.count(' '))

#se colocarmos so o strip para tirar os espaçoes inuteis, ele ainda vai contabilizar com o len o espaço que tem entre os nomes, como no caso debora mariano (conta o espaço
#entre o nome e o sobrenome), mas pra nao contar esses espaços, colocamos o -nome.count('').
```

```python
#d) Quantas letras tem o primeiro nome?

#Como a pessoa vai digitar o nome completo dela, primeiro temos que dividir o nome com o split, onde a primeira caixa que ele vai pegar é o primeiro nome,
#depois eu faço o print usando dentro do len que vai fazer a contagem (entao tenho que colocar o len) a variável dividindo porque nela colocamos o nome para dividir e
#com a informaçao que queremos a primeira caixinha, sendo ela a 0, porque em python a contagem começa em 0.

#obs: Veja que mesmo nao colocando o strip, ele nao contou os espaços que coloquei no inicio do meu nome. Isso acontece por causa que foi colocado o split, onde ele vai
#dividir em caixinhas o que tem letras.

nome = str(input('Digite seu nome completo: '))

dividindo = nome.split()

print(len(dividindo[0]))

#o split divide a frase em lista onde a primeira palavra ocupa a posicao 0, como queremos quantas letras tem o primeiro nome, so colocamos que queremos [0].
```

```python
#b)

frase = str(input('Digite uma frase: ')).strip()

minusculo = frase.lower()

print(minusculo.find('a')+1) #o +1 é pra informar que queremos saber qual foi a posicao que ela vai ter quando aparece a primeira vez, e como em python a numeraçao inicia em 0, ficaria diferente da nossa contagem normal.
```

```python
#c) Em que posiçao ela aparece a última vez?

frase = str(input('Digite uma frase: ')).strip()

minusculo = frase.lower()

print(minusculo.rfind('a')+1) #o rfind é para o nosso programa procurar o 'a' a partir do lado direito, onde de acordo com a nossa escrita, é quando terminamos de escrever.
```
